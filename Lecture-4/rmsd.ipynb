```python
import MDAnalysis as mda
from MDAnalysis.analysis import rms
import matplotlib.pyplot as plt

# --- Load the Universe ---
u = mda.Universe("rfc.psf", "rfc.dcd")

# --- Define the selection ---
selection = "backbone"
# MDAnalysis automatically aligns to your selection 

# --- Run the RMSD analysis ---
R = rms.RMSD(u, select=selection)
R.run()

# --- Plot results (Time is in index 0, RMSD is in index 2) ---
plt.plot(R.results.rmsd[:,0], R.results.rmsd[:,2], label="MDAnalysis")
plt.xlabel("Time (frames)")
plt.ylabel("RMSD (Å)")
plt.legend()
plt.show()
```

```python
import mdtraj as md

# --- Load the trajectory ---
traj = md.load("rfc.dcd", top="rfc.psf")

# --- MDTraj uses a 'select' method that returns an array of integers ---
backbone_indices = traj.topology.select("not hydrogen")

# --- Calculate RMSD ---
mdtraj_rmsd = md.rmsd(traj, traj, frame=0, atom_indices=backbone_indices)

# --- Plot ---
plt.plot(traj.time, mdtraj_rmsd, label="MDTraj", color='orange')
plt.xlabel("Step")
plt.ylabel("RMSD (nm)")
plt.legend()
plt.show()
```

```python
# --- This shows how the selection syntax can work ---
selection = "segid PROE and backbone"

# --- Run the RMSD analysis ---
R_PROE = rms.RMSD(u, select=selection)
R_PROE.run()

# --- Plot results (Time is in index 0, RMSD is in index 2) ---
plt.plot(R_PROE.results.rmsd[:,0], R_PROE.results.rmsd[:,2], label="MDAnalysis")
plt.xlabel("Time (frames)")
plt.ylabel("RMSD (Å)")
plt.legend()
plt.show()
```
